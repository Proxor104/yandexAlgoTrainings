# Problem A

## Solution

```python
def build_tree_recursive(name, relationships):
    children = [child for parent, child in relationships if parent == name]
    child_trees = [build_tree_recursive(child, relationships) for child in children]
    return [name, child_trees]

def calculate_heights(tree, height=0):
    name, children = tree
    heights = {name: height}
    for child in children:
        heights.update(calculate_heights(child, height + 1))
    return heights

def solutionA():
  dir = 'problemA/'
  input_name = dir+'input.txt'
  output_name = dir+'output.txt'
  with open(input_name, 'r') as input:
    lines = input.readlines()
    n = int(lines[0].strip())
    lines = lines[1:]
    #
    array = []
    children = set()
    for elem in lines:
      child = elem.strip().split()[0]
      parent = elem.strip().split()[1]
      children.add(child)
      array.append([parent, child])
    #
    root = ''
    for elem in array:
      if elem[0] not in children:
        root = elem[0]
        break
    tree = build_tree_recursive(root, array)
    heights = calculate_heights(tree)
    ret = []
    for name in sorted(heights):
      ret.append(str(name) + ' ' + str(heights[name]))
  with open(output_name, 'w') as output:
    output.write('\n'.join(ret))
```

```python
import sys
sys.setrecursionlimit(100000)

def build_tree(root, relationships, children, parents, height=0):
    child_trees = []
    if root in children.keys():
      for child in children[root]:
        parents[child] = (height + 1, root)
        tmp1, tmp2 = build_tree(child, relationships, children, parents, height + 1)
        child_trees.append(tmp1)
        parents.update(tmp2)
    return [root, child_trees], parents

def find_pair(parents, first, second):
  first_h = parents[first][0]
  second_h = parents[second][0]
  if second_h < first_h:
    first, second = second, first
    first_h, second_h = second_h, first_h
  #
  if first_h < second_h:
    while second_h > first_h:
      second = parents[second][1]
      second_h -= 1
  if second == first:
    return first
  height = second_h
  while height > 0 and second != first:
    first = parents[first][1]
    second = parents[second][1]
    height -= 1
  return first

def solutionC():
  dir = 'problemC/'
  input_name = dir+'input.txt'
  output_name = dir+'output.txt'
  with open(input_name, 'r') as input:
    lines = input.readlines()
    n = int(lines[0].strip())
    lines = lines[1:]
    #
    array = []
    members = set()
    set_children = set()
    children = {}
    i = 0
    while i < len(lines):
      if len(members) == n and len(set_children) == n-1:
        break
      elem = lines[i]
      child = elem.strip().split()[0]
      parent = elem.strip().split()[1]
      if parent not in children.keys():
        children[parent] = []
      children[parent].append(child)
      set_children.add(child)
      members.add(child)
      members.add(parent)
      array.append([parent, child])
      i += 1
    #
    root = (members - set_children).pop()
    #
    parents = {}
    ret = []
    tree = []
    tree, parents = build_tree(root, array, children, {}, 0)
    parents[root] = (0, None)
    #
    lines = lines[i:]
    ret = []
    for elem in lines:
      first = elem.strip().split()[0]
      second = elem.strip().split()[1]
      ret.append(find_pair(parents, first, second))
  with open(output_name, 'w') as output:
    output.write('\n'.join(ret)+'\n')
```

## Solution

```python
import sys
sys.setrecursionlimit(100000)

class Tree():
    def __init__(self):
      self.tree = []
      self.elements = set()
      self.string = ''
      self.is_changed = True

    def add(self, value):
      if value in self.elements:
        return 'ALREADY'
      else:
        self.is_changed = True
        return self.add_recursive(self.tree, value, 0)

    def add_recursive(self, subtree, value, height):
      if len(subtree) == 0:
        subtree.extend([(height, value), [], []])
        self.elements.add(value)
        return 'DONE'
      root, left, right = subtree
      if root[1] > value:
        return self.add_recursive(left, value, height + 1)
      else:
        return self.add_recursive(right, value, height + 1)

    def search(self, value):
      if value in self.elements:
        return 'YES'
      else:
        return 'NO'

    def printtree(self):
      if self.is_changed:
        self.string = ''
        self.is_changed = False
        self.printtree_recursive(self.tree)
      return self.string[:-1]

    def printtree_recursive(self, subtree):
      if len(subtree) == 0:
        return
      root, left, right = subtree
      self.printtree_recursive(left)
      ret = ['.'] * root[0]
      ret.append(str(root[1]))
      self.string += ''.join(ret) + '\n'
      self.printtree_recursive(right)



def solutionD():
  dir = 'problemD/'
  input_name = dir+'input.txt'
  output_name = dir+'output.txt'
  with open(input_name, 'r') as input:
    lines = input.readlines()
    n = len(lines)
    operations = [0] * n
    tree = Tree()
    ret = []
    for i, elem in enumerate(lines):
      tmp = elem.strip().split()
      if len(tmp) == 2 and tmp[0] == 'ADD':
        tmp = int(tmp[1])
        ret.append(tree.add(tmp))
      elif len(tmp) == 2 and tmp[0] == 'SEARCH':
        tmp = int(tmp[1])
        ret.append(tree.search(tmp))
      else:
        ret.append(tree.printtree())
  with open(output_name, 'w') as output:
    output.write('\n'.join(ret)+'\n')
```

# Problem E

## Solution

```python
import sys
sys.setrecursionlimit(100000)

def build_tree(root, relations, parents, num_children):
    num_children[root] = 0
    for child in relations[root]:
      if child not in parents:
        tmp = set()
        tmp.add(root)
        num_children = build_tree(child, relations, parents.union(tmp), num_children)
        num_children[root] += num_children[child] + 1
    return num_children

def solutionE():
  dir = 'problemE/'
  input_name = dir+'input.txt'
  output_name = dir+'output.txt'
  with open(input_name, 'r') as input:
    lines = input.readlines()
    n = int(lines[0].strip())
    lines = lines[1:]
    members = set()
    relations = {}
    for elem in lines:
      parent = int(elem.strip().split()[0])
      child = int(elem.strip().split()[1])
      if parent not in relations.keys():
        relations[parent] = []
      relations[parent].append(child)
      if child not in relations.keys():
        relations[child] = []
      relations[child].append(parent)
      members.add(child)
      members.add(parent)
    root = 1
    ret = [0] * n
    num_children = build_tree(root, relations, set(), {})
    for i in range(n):
      ret[i] = num_children[i+1] + 1
  with open(output_name, 'w') as output:
    output.write(' '.join(map(str, ret))+'\n')
```

# Problem G

```python
from collections import defaultdict, deque

def solve():
    input = """3 2 10
1 2
1 3"""
    data = input.strip().split()

    N, M, K = map(int, data[:3])
    edges = data[3:]

    graph = defaultdict(list)
    for i in range(M):
        a, b = map(int, edges[2 * i: 2 * i + 2])
        graph[a].append(b)
        graph[b].append(a)

    color = [-1] * (N + 1)  # -1 означает, что вершина не посещена

    def is_bipartite(start):
        queue = deque([start])
        color[start] = 0
        while queue:
            node = queue.popleft()
            for neighbor in graph[node]:
                if color[neighbor] == -1:  # Если сосед ещё не окрашен
                    color[neighbor] = 1 - color[node]
                    queue.append(neighbor)
                elif color[neighbor] == color[node]:  # Если окраска нарушена
                    return False
        return True

    components = 0
    for i in range(1, N + 1):
        if color[i] == -1:  # Если вершина не посещена
            if not is_bipartite(i):
                print(0)
                return
            components += 1

    # 2^(число компонент связности) * 2^(количество способов на компоненту)
    result = pow(2, components, K)
    print(result)

```

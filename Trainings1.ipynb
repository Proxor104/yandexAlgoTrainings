# Тестирующая система

```python
class Test():
  def __init__(self, solution, input_name = 'input.txt', output_name = 'output.txt',
               all_input_name = 'all_input.txt', all_output_name = 'all_output.txt',
               separator = '\n', answers = 'answers.txt', dir = './'):
    self.all_input_name = dir+all_input_name
    self.all_output_name = dir+all_output_name
    self.separator = separator
    self.input_name = dir+input_name
    self.output_name = dir+output_name
    self.solution = solution
    self.answers = dir+answers
  def test(self):
    with open(self.all_input_name, 'r+') as all_input:
      with open(self.all_output_name, 'w+') as all_output:
        data = []
        res = []
        for line in all_input:
          if line == self.separator:
            with open(self.input_name, 'w+') as input:
              input.writelines(data)
            data = []
            self.solution()
            with open(self.output_name, 'r+') as output:
              all_output.writelines(output.readlines())
              all_output.write('\n')
          else:
            data.append(line)
  def if_correct(self):
    with open(self.answers, 'r+') as ans:
      with open(self.all_output_name) as all_output:
        lines1 = ans.readlines()
        lines2 = all_output.readlines()
    arr = lines1.copy()
    for i, elem in enumerate(arr):
      lines1[i] = elem.strip()
      if elem == '\n':
        break
    arr = lines2.copy()
    for i, elem in enumerate(arr):
      lines2[i] = elem.strip()
      if elem == '\n':
        break
    print(lines1)
    print(lines2)
    print(lines1==lines2)
```

# Problem A

```python
def solutionA():
  dir = 'problemA/'
  input_name = dir+'input.txt'
  output_name = dir+'output.txt'
  with open(input_name, 'r') as input:
    lines = input.readlines()
    x1 = int(lines[0].strip())
    y1 = int(lines[1].strip())
    x2 = int(lines[2].strip())
    y2 = int(lines[3].strip())
    x = int(lines[4].strip())
    y = int(lines[5].strip())
    ret = ""
    #
    if y < y1:
      ret += 'S'
    elif y > y2:
      ret += 'N'
    #
    if x < x1:
      ret += 'W'
    elif x > x2:
      ret += 'E'
  with open(output_name, 'w') as output:
    output.write(ret)
```

```python
test = Test(dir = 'problemA/', solution = solutionA)
test.test()
test.if_correct()
```

# Problem B

```python
def solutionB():
  dir = 'problemB/'
  input_name = dir+'input.txt'
  output_name = dir+'output.txt'
  with open(input_name, 'r') as input:
    lines = input.readlines()
    A = int(lines[0].strip())
    B = int(lines[1].strip())
    C = int(lines[2].strip())
    D = int(lines[3].strip())
    arr = []
    if A > 0 and C > 0:
      arr.append([B+1, D+1])
    if B > 0 and D > 0:
      arr.append([A+1,C+1])
    if A > 0 and B > 0:
      arr.append([max(A, B) + 1, 1])
    if C > 0 and D > 0:
      arr.append([1, max(C,D) + 1])
    if len(arr) > 0:
      min = arr[0][0] + arr[0][1]
      ind = 0
      for i in range(1, len(arr)):
        tmp = arr[i][0] + arr[i][1]
        if tmp < min:
          min = tmp
          ind = i
    else:
      arr.append([0,0])
      ind = 0
    ret = str(arr[ind][0])+' '+str(arr[ind][1])
  with open(output_name, 'w') as output:
    output.write(ret)
```

```python
test = Test(dir = 'problemB/', solution = solutionB)
test.test()
test.if_correct()
```

```python
def print_image(image):
  for elem in image:
    print(''.join(elem))
  print()
```

```python
def del_left(arr):
  n = len(arr)
  if n > 0:
    m = len(arr[0])
  if n > 0 and m > 0:
    is_ok = True
    for i in range(n):
      is_ok &= arr[i][0] == '.'
    if is_ok:
      arr = [arr[i][1:] for i in range(n)]
      return 0, arr
    else:
      return -1, arr
  else:
    arr = []
    return -2, arr

def del_right(arr):
  n = len(arr)
  if n > 0:
    m = len(arr[0])
  if n > 0 and m > 0:
    is_ok = True
    for i in range(n):
      is_ok &= arr[i][m-1] == '.'
    if is_ok:
      arr = [arr[i][:m-1] for i in range(n)]
      return 0, arr
    else:
      return -1, arr
  else:
    arr = []
    return -2, arr

def del_up(arr):
  n = len(arr)
  if n > 0:
    m = len(arr[0])
  if n > 0 and m > 0:
    is_ok = True
    for j in range(m):
      is_ok &= arr[0][j] == '.'
    if is_ok:
      arr = arr[1:]
      return 0, arr
    else:
      return -1, arr
  else:
    arr = []
    return -2, arr

def del_down(arr):
  n = len(arr)
  if n > 0:
    m = len(arr[0])
  if n > 0 and m > 0:
    is_ok = True
    for j in range(m):
      is_ok &= arr[n-1][j] == '.'
    if is_ok:
      arr = arr[:n-1]
      return 0, arr
    else:
      return -1, arr
  else:
    arr = []
    return -2, arr

def srez(image):
  n = len(image)
  m = len(image[0])
  while n > 0 and m > 0:
    ret, image = del_left(image)
    if ret == -2:
      break
    #
    ret, image = del_right(image)
    if ret == -2:
      break
    #
    ret, image = del_up(image)
    if ret == -2:
      break
    #
    ret, image = del_down(image)
    if ret == -2:
      break
    #
    n_new = len(image)
    m_new = len(image[0])
    if n == n_new and m == m_new:
      break
    else:
      n = n_new
      m = m_new
  return image

def point_in_rec(i, j, left_up, right_down):
  return left_up[0] <= i <= right_down[0] and left_up[1] <= j <= right_down[1]

def point_in_recs(i, j, list_found):
  lezh = [point_in_rec(i, j, list_found[0][k], list_found[1][k]) for k in range(len(list_found[0]))]
  lezh.append(0)
  return max(lezh)

def find_rec(image, n, m, list_found, start_i, start_j):
  #
  lezh = point_in_recs(start_i, start_j, list_found)
  if not bool(lezh):
    list_found[0].append((start_i, start_j))
  else:
    return list_found, start_j
  #
  i_last = -1
  j_last = -1
  i = start_i
  while i < n:
    j = start_j
    tmp = image[i][start_j]
    if tmp != '.':
      if i_last == -1:
        i_last = i - 1
        list_found[1].append((i_last, j_last))
        return list_found, j_last
    j += 1
    #
    while j < m:
      tmp = image[i][j]
      if tmp != '.':
        if i_last == -1 and j_last == -1:
          j_last = j - 1
          j += 1
          break
        elif i_last == -1:
          j_last = min(j_last, j - 1)
          j += 1
          break
      j += 1
    tmp = image[i][j-1]
    if tmp == '.':
      j_last = j - 1
    i += 1
  if j_last == -1:
    j_last = max(j_last, m - 1)
  if i_last == -1:
    i_last = max(i_last, n - 1)
  list_found[1].append((i_last, j_last))
  return list_found, j_last

def make_list_found(image):
  n = len(image)
  m = 0
  if n > 0:
    m = len(image[0])
    if m > 0:
      list_found = [[], []]
      for i in range(n):
        j = 0
        while j < m:
          tmp = image[i][j]
          if tmp == '.':
            list_found, j = find_rec(image, n, m, list_found, i, j)
            if len(list_found[0]) > 2:
              return 'X'
          j += 1
      return list_found
  return 'X'

def solutionC():
  dir = 'problemC/'
  input_name = dir+'input.txt'
  output_name = dir+'output.txt'
  answer = ''
  with open(input_name, 'r') as input:
    lines = input.readlines()
    n = int(lines[0].strip())
    lines = lines[1:]
    image = []
    is_ok = len(lines) == n
    for line in lines:
      diods = list(line.strip())
      image.append(diods)
      is_ok &= (len(diods) == n)
    if is_ok:
      image = srez(image)
      ret = make_list_found(image)
      if type(ret) == str:
        answer = ret
      else:
        list_found = ret
        num_rec = len(list_found[0])
        n = len(image)
        m = len(image[0])
        if num_rec == 0:
          answer = 'I'
        elif num_rec == 1:
          y1 = list_found[0][0][0]
          y2 = list_found[1][0][0]
          x1 = list_found[0][0][1]
          x2 = list_found[1][0][1]
          if 0 < x1 <= x2 < m-1 and 0 < y1 <= y2 < n-1:
            answer = 'O'
          elif 0 < x1 <= x2 == m-1 and 0 < y1 <= y2 < n-1:
            answer = 'C'
          elif 0 < x1 <= x2 == m-1 and 0 == y1 <= y2 < n-1:
            answer = 'L'
          else:
            answer = 'X'
        elif num_rec == 2:
          y1 = list_found[0][1][0]
          x1 = list_found[0][1][1]
          y2 = list_found[1][1][0]
          x2 = list_found[1][1][1]
          y3 = list_found[0][0][0]
          x3 = list_found[0][0][1]
          y4 = list_found[1][0][0]
          x4 = list_found[1][0][1]
          if 0 < x1 == x3 <= x2 == x4 < m-1 and 0 == y3 <= y4 < y1 <= y2 == n-1:
            answer = 'H'
          elif 0 < x1 == x3 <= x4 < x2 == m-1 and 0 < y3 <= y4 < y1 <= y2 == n-1:
            answer = 'P'
          else:
            answer = 'X'
    else:
      print("Something went wrong")
  with open(output_name, 'w') as output:
    output.write(answer)
```

```python
test = Test(dir = 'problemC/', solution = solutionC)
test.test()
test.if_correct()
```

```python
def solutionD():
  dir = 'problemD/'
  input_name = dir+'input.txt'
  output_name = dir+'output.txt'
  with open(input_name, 'r') as input:
    line = input.readlines()[0].strip().split(' ')
    n = int(line[0])
    m = int(line[1])
    mode = (line[2])
    answer = ''
    if mode == "heat" and n < m:
      print(1)
      answer = m
    elif mode == "heat" and n >= m:
      print(2)
      answer = n
    elif mode == "freeze" and n > m:
      print(3)
      answer = m
    elif mode == "freeze" and n <= m:
      print(4)
      answer = n
    elif mode == "auto":
      print(5)
      answer = m
    elif mode == "fan":
      print(6)
      answer = n

  with open(output_name, 'w') as output:
    output.write(str(answer))
```

```python
class Test():
  def __init__(self, solution, input_name = 'input.txt', output_name = 'output.txt',
               all_input_name = 'all_input.txt', all_output_name = 'all_output.txt',
               separator = '\n', answers = 'answers.txt', dir = './'):
    self.all_input_name = dir+all_input_name
    self.all_output_name = dir+all_output_name
    self.separator = separator
    self.input_name = dir+input_name
    self.output_name = dir+output_name
    self.solution = solution
    self.answers = dir+answers
  def test(self, num_tests = -1):
    if num_tests == -1:
      with open(self.all_input_name, 'r+') as all_input:
        with open(self.all_output_name, 'w+') as all_output:
          data = []
          res = []
          for line in all_input:
            if line == self.separator:
              with open(self.input_name, 'w+') as input:
                input.writelines(data)
              data = []
              self.solution()
              with open(self.output_name, 'r+') as output:
                all_output.writelines(output.readlines())
                all_output.write('\n')
            else:
              data.append(line)
    else:
      with open(self.all_input_name, 'r+') as all_input:
        with open(self.all_output_name, 'w+') as all_output:
          data = []
          res = []
          count = 0
          for line in all_input:
            if count < num_tests:
              if line == self.separator:
                count += 1
                with open(self.input_name, 'w+') as input:
                  input.writelines(data)
                data = []
                self.solution()
                with open(self.output_name, 'r+') as output:
                  all_output.writelines(output.readlines())
                  all_output.write('\n')
              else:
                data.append(line)
            else:
              break
  def if_correct(self, num_tests = -1):
    with open(self.answers, 'r+') as ans:
      with open(self.all_output_name) as all_output:
        lines1 = ans.readlines()
        lines2 = all_output.readlines()
    arr = lines1.copy()
    for i, elem in enumerate(arr):
      lines1[i] = elem.strip()
      if elem == '\n':
        break
    arr = lines2.copy()
    for i, elem in enumerate(arr):
      lines2[i] = elem.strip()
      if elem == '\n':
        break
    if num_tests != -1:
      lines1 = lines1[:num_tests]
    print(lines1)
    print(lines2)
    if lines1==lines2:
      print(True)
    else:
      i = 0
      for elem1, elem2 in zip(lines1, lines2):
        if elem1 != elem2:
          print(f"wrong: {i} : {elem1}, {elem2}")
        i += 1
```

```python
from collections import deque
def solutionC():
  dir = 'problemC/'
  input_name = dir+'input.txt'
  output_name = dir+'output.txt'
  with open(input_name, 'r') as input:
    lines = input.readlines()
    n, k = list(map(int, lines[0].strip().split()))
    array = list(map(int, lines[1].strip().split()))
    ret = [0] * (n - k + 1)
    d = deque()
    window_size = 0
    for i, elem in enumerate(array):
      if window_size == k:
        ret[i - k] = d[0]
        if array[i - k] == d[0]:
          d.popleft()
        window_size -= 1
      while len(d) > 0 and elem < d[-1]:
        d.pop()
      d.append(elem)
      window_size += 1
    ret[n - k] = d[0]
  with open(output_name, 'w') as output:
    output.write(' '.join(list(map(str, ret))))
```

```python
def make_postfix(array):
  stack =[]
  ret = ''
  priorities = {'+': 0, '-': 0, '*': 1}
  is_oper = lambda x: x == '*' or x == '+' or x == '-'
  count_opers = 0
  count_nums = 0
  for elem in array:
    if elem.isdigit():
      ret += ' ' + elem
      count_nums += 1
    elif is_oper(elem):
      while len(stack) > 0 and is_oper(stack[-1]) and priorities[stack[-1]] >= priorities[elem]:
        ret += ' ' + stack.pop()
        count_opers += 1
      stack.append(elem)
    elif elem == '(':
      stack.append(elem)
    elif elem == ')':
      while len(stack) > 0 and stack[-1] != '(':
        ret += ' ' + stack.pop()
        count_opers += 1
      if len(stack) == 0:
        ret = 'WRONG'
        break
      stack.pop()
    else:
      ret = 'WRONG'
      break
  while len(stack) > 0 and is_oper(stack[-1]):
    ret += ' ' + stack.pop()
    count_opers += 1
  if (count_opers != count_nums - 1):
    ret = 'WRONG'
  return ret


def compute(array):
  stack = []
  is_num = lambda x: x != '*' and x != '+' and x != '-'
  for i, elem in enumerate(array):
    if is_num(elem):
      elem = int(elem)
      stack.append(elem)
    else:
      if elem == '*':
        tmp = stack[-1] * stack[-2]
      elif elem == '+':
        tmp = stack[-1] + stack[-2]
      elif elem == '-':
        tmp = stack[-2] - stack[-1]
      stack.pop()
      stack.pop()
      stack.append(tmp)
  return stack.pop()

def solutionE():
  dir = 'problemE/'
  input_name = dir+'input.txt'
  output_name = dir+'output.txt'
  with open(input_name, 'r') as input:
    lines = input.readlines()
    array = lines[0].strip().strip(' ')
    ret = 0
    # предобработка
    array_in = []
    num = ''
    flag = False
    for elem in array:
      if elem != ' ':
        if elem == '+' or elem == '-' or elem == '*' or elem == '(' or elem == ')':
          if flag:
            array_in.append(num)
            num = ''
            flag = False
          array_in.append(elem)
        elif elem.isdigit():
          if not flag:
            flag = True
          num += elem
        else:
          ret = -1
          break
      else:
        if flag:
          array_in.append(num)
          num = ''
          flag = False
    if flag:
      array_in.append(num)
      num = ''
      flag = False
    #
    if ret == -1:
      ret = 'WRONG'
    else:
      # перевод в постфиксную запись
      ret = make_postfix(array_in)
      if ret != 'WRONG':
        ret = compute(ret.split())
  with open(output_name, 'w') as output:
    output.write(str(ret))
```

```python
def solutionF():
  dir = 'problemF/'
  input_name = dir+'input.txt'
  output_name = dir+'output.txt'
  with open(input_name, 'r') as input:
    lines = input.readlines()
    n = int(lines[0].strip())
    w = list(lines[1].strip())
    if len(lines) == 3:
      s = list(lines[2].strip())
    else:
      s = []
    ret = ''.join(s)
    stack = []
    is_open = lambda x: x == '(' or x == '['
    is_close = lambda x: x == ')' or x == ']'
    ret_square_circle = lambda x : ']' if x == '[' else ')'
    min_open = ''
    for elem in w:
      if elem == '(' or elem == '[':
        min_open = elem
        break
    dict_w = {}
    for i, elem in enumerate(w):
      dict_w[elem] = i
    #
    for elem in s:
      if is_open(elem):
        stack.append(elem)
      else:
        stack.pop()
    #
    len_s = len(stack)
    available = int((n - len_s - len(s)) / 2)
    while available > 0:
      if len(stack) != 0:
        tmp = ret_square_circle(stack[-1])
        if dict_w[min_open] > dict_w[tmp]:
          ret += tmp
          stack.pop()
          continue
      ret += min_open
      stack.append(min_open)
      available -= 1
    while len(stack) > 0:
      tmp = ret_square_circle(stack[-1])
      ret += tmp
      stack.pop()
  with open(output_name, 'w') as output:
    output.write(str(ret))
```

```python
from collections import deque
def solutionG():
  dir = 'problemG/'
  input_name = dir+'input.txt'
  output_name = dir+'output.txt'
  with open(input_name, 'r') as input:
    lines = input.readlines()
    n, b = list(map(int, lines[0].strip().split()))
    array = list(map(int, lines[1].strip().split()))
    d = deque()
    sum = 0
    for i in range(len(array)):
      d.appendleft([array[i], i])
      sub_value = b
      while len(d) > 0 and sub_value > 0:
        if sub_value >= d[-1][0]:
          sub_value -= d[-1][0]
          sum += d[-1][0] * (i - d[-1][1] + 1)
          d.pop()
        else:
          d[-1][0] -= sub_value
          sum += sub_value * (i - d[-1][1] + 1)
          sub_value = 0
    i = len(array)
    for elem in d:
      sum += elem[0] * (i - elem[1] + 1)
    ret = sum
  with open(output_name, 'w') as output:
    output.write(str(ret))
```

# Problem H

## Solution

```python
class Stack():
  def __init__(self):
    self.stack = []
    self.prefix_sum = [0]
    self.len = 0

  def push(self, x):
    self.prefix_sum.append(self.prefix_sum[self.len] + x)
    self.len += 1
    return self.stack.append(x)

  def pop(self):
    self.prefix_sum.pop()
    self.len -= 1
    return self.stack.pop()

  def get_sum(self, k):
    return self.prefix_sum[self.len] - self.prefix_sum[self.len - k]

def solutionH():
  dir = 'problemH/'
  input_name = dir+'input.txt'
  output_name = dir+'output.txt'
  with open(input_name, 'r') as input:
    lines = input.readlines()
    n = int(lines[0].strip())
    array = [''] * n
    s = Stack()
    ret = []
    for i in range(n):
      array[i] = str(lines[i+1].strip())
      if array[i] == '-':
        ret.append(s.pop())
      else:
        oper = array[i][0]
        if oper == '+':
          s.push(int(array[i][1:]))
        elif oper == '?':
          ret.append(s.get_sum(int(array[i][1:])))
  with open(output_name, 'w') as output:
    output.write(' '.join(list(map(str, ret))))
```

```python
from collections import deque
def solutionI():
  dir = 'problemI/'
  input_name = dir+'input.txt'
  output_name = dir+'output.txt'
  with open(input_name, 'r') as input:
    lines = input.readlines()
    n = int(lines[0].strip())
    mainRoad1, mainRoad2 = list(map(int, lines[1].strip().split()))
    mainRoad1, mainRoad2 = mainRoad1 - 1, mainRoad2 - 1
    #
    together, right, left = False, False, False
    if (mainRoad1 - mainRoad2) % 2 == 0:
      together = True
      spin = min(mainRoad2, mainRoad1)
    else:
      together = False
      spin = min(mainRoad2, mainRoad1)
      mainRoad1 = (mainRoad1 - spin) % 4
      mainRoad2 = (mainRoad2 - spin) % 4
      if max(mainRoad1, mainRoad2) == 1:
        right = True
        left = False
      elif max(mainRoad1, mainRoad2) == 3:
        right = False
        left = True
    #
    array = [0] * n
    for i in range(n):
      array[i] = list(map(int, lines[i+2].strip().split()))
      array[i].append(i)
      array[i][0] -= 1
      array[i][1] -= 1
      if together:
        array[i][0] = (array[i][0] - spin) % 4
      if not together:
        array[i][0] = (array[i][0] - spin) % 4
    ret = [0] * n
    array.sort(key=lambda x : x[1])
    #
    d = [deque(), deque(), deque(), deque()]
    length = [0, 0, 0, 0]
    for elem in array:
      d[elem[0]].appendleft(elem)
      length[elem[0]] += 1
    i = 0
    while max(length) > 0:
      # if main road is straight
      if together:
        bool1 = len(d[0]) > 0 and d[0][-1][1] <= i
        bool2 = len(d[2]) > 0 and d[2][-1][1] <= i
        if bool1 or bool2:
          if bool1:
            ret[d[0][-1][2]] = i + 1
            d[0].pop()
            length[0] -= 1
          if bool2:
            ret[d[2][-1][2]] = i + 1
            d[2].pop()
            length[2] -= 1
        else:
          if len(d[1]) > 0 and d[1][-1][1] <= i:
            ret[d[1][-1][2]] = i + 1
            d[1].pop()
            length[1] -= 1
          if len(d[3]) > 0 and d[3][-1][1] <= i:
            ret[d[3][-1][2]] = i + 1
            d[3].pop()
            length[3] -= 1
      # if main road leads from the north to the east
      elif right:
        if len(d[0]) > 0 and d[0][-1][1] <= i:
          ret[d[0][-1][2]] = i + 1
          d[0].pop()
          length[0] -= 1
        elif len(d[1]) > 0 and d[1][-1][1] <= i:
          ret[d[1][-1][2]] = i + 1
          d[1].pop()
          length[1] -= 1
        elif len(d[2]) > 0 and d[2][-1][1] <= i:
          ret[d[2][-1][2]] = i + 1
          d[2].pop()
          length[2] -= 1
        elif len(d[3]) > 0 and d[3][-1][1] <= i:
          ret[d[3][-1][2]] = i + 1
          d[3].pop()
          length[3] -= 1
      # if main road leads from the north to the west
      elif left:
        if len(d[3]) > 0 and d[3][-1][1] <= i:
          ret[d[3][-1][2]] = i + 1
          d[3].pop()
          length[3] -= 1
        elif len(d[0]) > 0 and d[0][-1][1] <= i:
          ret[d[0][-1][2]] = i + 1
          d[0].pop()
          length[0] -= 1
        elif len(d[1]) > 0 and d[1][-1][1] <= i:
          ret[d[1][-1][2]] = i + 1
          d[1].pop()
          length[1] -= 1
        elif len(d[2]) > 0 and d[2][-1][1] <= i:
          ret[d[2][-1][2]] = i + 1
          d[2].pop()
          length[2] -= 1
      i += 1
  with open(output_name, 'w') as output:
    output.write(' '.join(list(map(str, ret))))
```
